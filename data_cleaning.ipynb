```python
import pandas as pd
import numpy as np
import datetime as dt
import glob
import os
```

```python
path = 'Resources/BikeTripHistories/'                     
all_files = glob.glob(os.path.join(path, "*.csv"))
bike_trips = pd.concat([pd.read_csv(f) for f in all_files], ignore_index=True)
```

```python
path = 'Resources/RidershipandMembershipData/'                     
all_files = glob.glob(os.path.join(path, "*.csv"))
riders = pd.concat([pd.read_csv(f) for f in all_files], ignore_index=True)
```

```python
riders.head()
```

```python
riders.count()
```

```python
#Split a column and add new date/time columns to existing dataframe
bike_trips[["start date", "start time"]] = bike_trips["starttime"].str.split(" ", n = 1, expand = True) 
  
# Dropping old starttime column
bike_trips.drop(columns =["starttime"], inplace = True) 

#Split a column and add new date/time columns to existing dataframe
bike_trips[["stop date", "stop time"]] = bike_trips["stoptime"].str.split(" ", n = 1, expand = True) 
  
# Dropping old stoptime column 
bike_trips.drop(columns =["stoptime"], inplace = True)

# df display 
bike_trips.head()
```

```python
# convert both date columns to datetime objects
bike_trips["start date"] = pd.to_datetime(bike_trips["start date"])
bike_trips["stop date"] = pd.to_datetime(bike_trips["stop date"])
```

```python
#Remove milliseconds from both time columns
bike_trips["start time"] = bike_trips["start time"].astype(str).str[:-5]
bike_trips["stop time"] = bike_trips["stop time"].astype(str).str[:-5]

#Convert both time columns to time objects
bike_trips["start time"] = pd.to_datetime(bike_trips["start time"]).dt.time
bike_trips["stop time"] = pd.to_datetime(bike_trips["stop time"]).dt.time
```

```python
bike_trips.to_csv("CleanData/bike_trips.csv")
riders.to_csv("CleanData/riders.csv")
```
